```python
from google.colab import userdata
GROQ_API_KEY = userdata.get('GROQ_API_KEY')
WEATHERSTACK_API_KEY = userdata.get('WEATHERSTACK_API_KEY')
```

```python
from langchain_groq import ChatGroq
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun

search_tool = DuckDuckGoSearchRun()
```

```python
@tool
def get_weather_data(city: str) -> str:
    """
    This tool fetches the current weather data for a given city
    """
    url = f'https://api.weatherstack.com/current?access_key=07a3b48e4c982f27e0e630fd4ea2b0a1={city}'
    response = requests.get(url)
    return response.json()
```

```python
# llm = ChatGroq(model="openai/gpt-oss-20b", temperature=0)
llm = ChatGroq(
    model="openai/gpt-oss-20b",
    temperature=0,
    #enable tool calling for this model
    model_kwargs={"tool_choice": "auto"},
    api_key=GROQ_API_KEY

)
```

```python
from langchain.agents import create_agent
from langsmith import client
```

```python
from langchain.agents.middleware import dynamic_prompt, ModelRequest
```

```python
client=client.Client()
prompt = client.pull_prompt("hwchase17/react") # pulls the standard react agent prompt
prompt_template_string = prompt.template #Extract the template string from the PromptTemplate object
```

```python
agent = create_agent(
    llm,
    tools=[search_tool, get_weather_data]
)
```

```python
response = agent.invoke(
    {"messages": [{"role": "user", "content": "Find the capital of Madhya Pradesh, then find it's current weather condition"}]}
)
print(response)
```

```python
for messages in response['messages']:
  print(messages)
```

```python
import gradio as gr
```

```python
def get_response(query):
    response = agent.invoke(
        {"messages": [{"role": "user", "content": query}]}
    )
    # Access the content attribute using .content instead of ['content']
    return response['messages'][-1].content
```

```python
iface = gr.Interface(
    fn=get_response,
    inputs=gr.Textbox(
        label="Ask a question to your AI Agent",
        placeholder="e.g., Find the the capital of West Bengal, then find it's current weather condition",
        lines=2
    ),
    outputs=gr.Textbox(label="Response", lines=10),
    title="AI Agent with Web Access",
    description="This AI Agent has access to the internet. You can ask me anything and it will search the web",
    examples=[
        ["Differentiate between Vectors and Vector store"],
        ["What is RAG model"],
        ["What is the current market cap of NVIDIA"]
    ]
)
```

```python
iface.launch(share= True)
```
